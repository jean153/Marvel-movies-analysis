```python
def box_office_totals(mcu_movies):
    phase_count = int(input("How many phases do you want to enter? "))
    results = []

    for i in range(phase_count):
        user_input = input(f"Enter Marvel Phase: ")
        phase = user_input.capitalize()

        phase_movies = mcu_movies[mcu_movies['Phase'] == phase]

        if phase_movies.empty:
            print(f"No data found for Phase {phase}.")
            results.append((phase, 0))
        else:
            total_box_office = phase_movies['Total Box Office(inflation)'].sum()
            results.append((phase, total_box_office))

    return results

# Now call the function and print the results
totals = box_office_totals(mcu_movies)

for phase, boxoffice in totals:
    print(f"Total box office for Phase {phase}: ${boxoffice:,.2f}")
```

```python
results = box_office_totals(mcu_movies)
fig, ax=plt.subplots()
phases=[phase for phase,total in results]
y_pos=np.arange(len(phases))
box_office_totals=[total for phase,total in results]
bar_colors=['tab:red','tab:blue','tab:green','#FFD700','tab:purple']
ax.barh(y_pos, box_office_totals, align='center',color=bar_colors[:len(phases)])
ax.set_yticks(y_pos, labels=phases)
ax.invert_yaxis()
ax.set_xlabel('Box office totals/phase')
ax.set_title('Box ofice total for each phase')
plt.savefig('phases.png')
```

```python
#line chart to show frequency of release months 
release_months= pd.DataFrame({
    "Movie":mcu['Movie'],
    "Release Month":mcu['Release Month'],
})
```

```python
release_months=release_months.loc[0:32]
```

```python
release_months
```

```python
months = pd.to_datetime(release_months['Release Month'], format='%B', errors='coerce')
release_months['Month_Name'] = months.dt.strftime('%B')
frequency=release_months.drop(['Release Month'], axis=1)
grouped_data = frequency.groupby('Month_Name')

```

```python
grouped_data.count()
```

```python
grouped_data = frequency.groupby(['Month_Name'])['Movie'].count().reset_index()
```
